```python
covid.drop(['Sno', 'ConfirmedIndianNational', 'ConfirmedForeignNational'], inplace = True, axis = 1, errors='ignore')
```

```python
covid.head()
```

```python
covid['Active_Cases'] = covid['Confirmed']-(covid['Cured']+covid['Deaths'])
covid.head()
```

```python
statewise = pd.pivot_table(covid,values=['Confirmed','Deaths','Cured'],index = 'State/UnionTerritory',aggfunc=max)
```

```python
statewise['Recovery Rate'] = statewise['Cured']*100/statewise['Confirmed']
```

```python
statewise['Mortality Rate'] = statewise['Deaths']*100/statewise['Confirmed']
```

```python
statewise = statewise.sort_values(by ='Confirmed',ascending=False)
```

```python
statewise.style.background_gradient(cmap='cubehelix')
```

```python
top_10 = covid.groupby(by = 'State/UnionTerritory').max()[['Active_Cases','Time']].sort_values(by = ['Active_Cases'],ascending=False)
```

```python
fig = plt.figure(figsize=(16,9))
```

```python
plt.title('Top states with active cases' , size = 25)
```

```python
ax = sns.barplot(data =top_10.iloc[:10], y ='Active_Cases',x = 'State/UnionTerritory',linewidth = 2, edgecolor ='red' )
```

```python
top_10 = covid.groupby(by = 'State/UnionTerritory').max()[['Active_Cases','Time']].sort_values(by = ['Active_Cases'],ascending=False)
fig = plt.figure(figsize=(16,9))
plt.title('Top states with active cases' , size = 25)
ax = sns.barplot(data =top_10.iloc[:10], y ='Active_Cases',x = 'State/UnionTerritory',linewidth = 2, edgecolor ='red' )
plt.xlabel = 'States'
plt.ylabel = 'Total Active Cases'
plt.show()
```

```python
top_10_deaths = covid.groupby(by ='State/UnionTerritory').max()[['Deaths','Time']].sort_values(by =['Deaths'],ascending = False).reset_index()
fig = plt.figure(figsize=(18,5))
plt.title('Top 10 deaths', size = 25)
ax = sns.barplot(data = top_10_deaths.iloc[:12], y ='Deaths' , x = 'State/UnionTerritory', linewidth=2, edgecolor = 'red')
ax.set_xlabel('States')
ax.set_ylabel('Deaths cases')
plt.show()
```

```python
covid['Time'] = pd.to_datetime(covid['Time'])
fig = plt.figure(figsize=(12,6))
ax = sns.lineplot(data = covid[covid['State/UnionTerritory'].isin(['Maharashtra','Karnataka','Kerala','Tamil Nadu','Uttar Pradesh'])], x = 'Time' , y = 'Active_Cases', hue = 'State/UnionTerritory')
ax.set_title('Top 5 affected states')
plt.show()
```

```python
vac = vaccine[vaccine.State !='India']
vac
```

```python
vac.rename(columns = {'Total Individuals Vaccinated':'Total'}, inplace = True)
vac.head()
```

```python
max_vac = vac.groupby('State')['Total'].sum().to_frame('Total')
max_vac = max_vac.sort_values('Total',ascending=False)[:5]
max_vac
```

```python
fig = plt.figure(figsize=(10,5))
plt.title('Top vaccinated state')
max_vac_plot = max_vac.reset_index()
x = sns.barplot(data = max_vac_plot.iloc[:10], y ='Total', x = 'State' , linewidth = 2, edgecolor ='red')
x.set_xlabel('States')
x.set_ylabel('Vaccinated')
plt.show()
```
